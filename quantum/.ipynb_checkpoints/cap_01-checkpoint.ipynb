```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
```

```python
#CONVERT CARTESIAN TO POLAR
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```

# Operaciones Basicas

```python
c1 = 4 + 5j
c2 = 6 - 6j
c3 = -4 + 10j

i_a = 0 + 0j
i_m = 1 + 0j

#OPERACIONES BASICAS
print('Suma: ', c1 + c2)
print('Resta: ', c1 - c2)
print('Multiplicacion: ', c1 * c2)
print('Division: ', c1 / c2)
print('')

#MODULO
print('Modulos: ', abs(3 + 4j), abs(3 - 4j), abs(-3 + 4j), abs(-3 - 4j))
print('Multiplicacion de modulos: ', abs(c1) * abs(c2), abs(c1 * c2))

#CAMBIO DE SIGNOS
print('Cambio en los dos signos: ', c1, c1 * -1)
print('Complejo conjugado: ', c1, c1.conjugate())
print('')

#CONJUGADOS
print('Conjugado mantiene adicion: ', c1.conjugate() + c2.conjugate(), (c1 + c2).conjugate())
print('Conjugado mantiene multiplicacion: ', c1.conjugate() * c2.conjugate(), (c1 * c2).conjugate())
print('Modulo al cuadrado: ', abs(c2) ** 2, c2 * c2.conjugate())
```

# Asociatividad y conmutatividad

```python
#ASOCIATIVIDAD
print('Asociatividad')
print(c1 + c2, c2 + c1)
print(c2 * c1, c2 * c1)
print('')

#CONMUTATIVIDAD
print('Conmutatividad')
print((c1 + c2) + c3, c1 + (c2 + c3))
print((c1 * c2) * c3, c1 * (c2 * c3))
print('')

#DISTRIBUCION DE MULTIPLICACION SOBRE SUMA
print('Distribucion')
print(c1 * (c2 + c3), c1 * c2 + c1 * c3)
print('')

#IDENTIDAD ADITIVA Y MULTIPLICATIVA
print('Identidad aditiva: ', c1, c1 + i_a)
print('Identidad multiplicativa: ', c1, c1 * i_m)
```

# Coordenas Polares


```python
magnitud, angulo = cart2pol(3, 4)
print('Magnitud y angulo: ', magnitud, angulo * (180 / np.pi))

x, y = pol2cart(magnitud, angulo)
print('Cartesianas: ', x, y)

print('')

magnitud, angulo = cart2pol(-3, 4)
print('Magnitud y angulo: ', magnitud, angulo * (180 / np.pi))

x, y = pol2cart(magnitud, angulo)
print('Cartesianas: ', x, y)

print('')

magnitud, angulo = cart2pol(-3, -4)
print('Magnitud y angulo: ', magnitud, angulo * (180 / np.pi) % 360)

x, y = pol2cart(magnitud, angulo)
print('Cartesianas: ', x, y)

print('')

magnitud, angulo = cart2pol(3, -4)
print('Magnitud y angulo: ', magnitud, angulo * (180 / np.pi) % 360)

x, y = pol2cart(magnitud, angulo)
print('Cartesianas: ', x, y)
```

```python
#ROTACION COMO MULTIPLICACION
c1 = 4 + 0j
c2 = c1 * 1j
c3 = c1 * 1j * 1j
c4 = c1 * 1j * 1j * 1j

plt.figure(figsize=(2, 2))
plt.scatter(c1.real, c1.imag, c='blue')
plt.scatter(c2.real, c2.imag, c='green')
plt.scatter(c3.real, c3.imag, c='yellow')
plt.scatter(c4.real, c4.imag, c='red')
plt.show()
```

```python
#AMPLIFICACION COMO MULTIPLICACION
c1 = 4 + 1j
c2 = c1 * 2
c3 = c1 * 3
c4 = c1 * 4

plt.figure(figsize=(2, 2))
plt.scatter(c1.real, c1.imag, c='blue')
plt.scatter(c2.real, c2.imag, c='green')
plt.scatter(c3.real, c3.imag, c='yellow')
plt.scatter(c4.real, c4.imag, c='red')
plt.show()
```

```python
#AMPLIFICACION COMO MULTIPLICACION
f = 2 + 1.5j
c1 = 4 + 1j
c2 = c1 * f
c3 = c1 * f * f
c4 = c1 * f * f * f

plt.figure(figsize=(2, 2))
plt.scatter(c1.real, c1.imag, c='blue')
plt.scatter(c2.real, c2.imag, c='green')
plt.scatter(c3.real, c3.imag, c='yellow')
plt.scatter(c4.real, c4.imag, c='red')
plt.show()
```

```python
#AMPLIFICACION COMO POTENCIA
c1 = 2.5 - 2j
c2 = c1 ** 2
c3 = c1 ** 3
c4 = c1 ** 4

plt.figure(figsize=(2, 2))
plt.scatter(c1.real, c1.imag, c='blue')
plt.scatter(c2.real, c2.imag, c='green')
plt.scatter(c3.real, c3.imag, c='yellow')
plt.scatter(c4.real, c4.imag, c='red')
plt.show()
```

```python
#RAICES DE 1 + FASE * j
array_x = []
array_y = []

magnitud = 1
fase = np.pi
roots = 40
for k in range(0, roots):
    
    angulo = (1 / roots) *  (fase + k * 2 * np.pi)

    x, y = pol2cart(magnitud, angulo)
    array_x.append(x)
    array_y.append(y)

plt.figure(figsize=(4, 4))
plt.scatter(array_x, array_y, c='blue')
plt.show()
```

# Representacion de Euler

```python
#COSENO
x = np.linspace(0, 10, 1000)
y = np.cos(x)

aprox_taylor = np.linspace(0, 0, 1000)

for n in range(0, 20):

    aprox_i = ((-1) ** n) * (x ** (2 * n)) / math.factorial(2 * n)
    aprox_taylor = aprox_taylor + aprox_i

plt.figure(figsize=(12, 2))
plt.plot(x, y, c='blue')
plt.plot(x, aprox_taylor, c='red')
plt.show()
```

```python
#SENO
x = np.linspace(0, 10, 1000)
y = np.sin(x)

aprox_taylor = np.linspace(0, 0, 1000)

for n in range(0, 20):

    aprox_i = ((-1) ** n) * (x ** (2 * n + 1)) / math.factorial(2 * n + 1)
    aprox_taylor = aprox_taylor + aprox_i

plt.figure(figsize=(12, 2))
plt.plot(x, y, c='blue')
plt.plot(x, aprox_taylor, c='red')
plt.show()
```

```python
#EXPONENCIAL
x = np.linspace(0, 5, 1000)
y = np.exp(x)

aprox_taylor = np.linspace(0, 0, 1000)

for n in range(0, 20):

    aprox_i = (x ** n) / math.factorial(n)
    aprox_taylor = aprox_taylor + aprox_i

plt.figure(figsize=(12, 2))
plt.plot(x, y, c='blue')
plt.plot(x, aprox_taylor, c='red')
plt.show()
```

```python
#RAICES DE 1 CON FORMA EXPONENCIAL
array_x = []
array_y = []

magnitud = 1
fase = np.pi
roots = 20
for k in range(0, roots):
    
    root = np.exp(2 * np.pi * 1j * k / roots)

    array_x.append(root.real)
    array_y.append(root.imag)

plt.figure(figsize=(4, 4))
plt.scatter(array_x, array_y, c='blue')
plt.show()
```
