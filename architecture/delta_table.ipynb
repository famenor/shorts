```python
import pandas as pd
from deltalake import DeltaTable, write_deltalake
```

```python
df = pd.DataFrame({"id": [1, 2, 3], 
                   "name": ["Aadhya", "Bob", "Chen"]
                  })

print(df.shape)
display(df)
```

```python
path = '/home/armando/git/shorts/architecture/data/deltalake/demo'

write_deltalake(table_or_uri=path, data=df, mode='overwrite', name='people_table', description='A simple people table')
```

```python
df_02 = pd.DataFrame({"id": [4, 5],
                      "name": ["Diana", "Eli"]
                       })

display(df_02)
```

```python
path = '/home/armando/git/shorts/architecture/data/deltalake/demo'

write_deltalake(table_or_uri=path, data=df_02, mode='append', name='people_table', description='A simple people table')
```

```python
dt = DeltaTable(path)

updated_data = pd.DataFrame({"id": [5, 6],
                             "name": ["Elizabeth", "Fabio"]
                            })

dt.merge(source=updated_data, predicate="target.id = source.id",
         source_alias="source", target_alias="target").when_matched_update_all().when_not_matched_insert_all().execute()
```
