```python
import pandas as pd
from deltalake import DeltaTable, write_deltalake
```

```python
df = pd.DataFrame({"id": [1, 2, 3], 
                   "name": ["Aadhya", "Bob", "Chen"]
                  })

print(df.shape)
display(df)
```

```python
path = '/home/armando/git/shorts/architecture/data/deltalake/demo'

write_deltalake(table_or_uri=path, data=df, mode='overwrite')
```
